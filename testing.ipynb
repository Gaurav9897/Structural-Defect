```python
import google.generativeai as genai
import os
```

```python
from PIL import Image
```

```python
defect_image = Image.open('defect.jpg')
```

```python
gemini_api_key = os.getenv('GOOGLE_API_KEY1')
genai.configure(api_key=gemini_api_key)
model = genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt = 'Identify the structural defect in the image and give solution'
```

```python
response = model.generate_content([prompt,defect_image],generation_config={'temperature':0.7})
```

```python
print(response.text)
```
